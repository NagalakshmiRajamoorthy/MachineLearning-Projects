```python
#Transform the values to numbers
from sklearn.preprocessing import LabelEncoder
b=LabelEncoder()
a["Gender"]=b.fit_transform(a["Gender"])
a["Gender"]=b.fit_transform(a["Gender"])
a["family_history"]=b.fit_transform(a["family_history"])
a["FAVC"]=b.fit_transform(a["FAVC"])
a["CAEC"]=b.fit_transform(a["CAEC"])
a["SMOKE"]=b.fit_transform(a["SMOKE"])
a["SCC"]=b.fit_transform(a["SCC"])
a["CALC"]=b.fit_transform(a["CALC"])
a["Obesity"]=b.fit_transform(a["Obesity"])
a["MTRANS"]=b.fit_transform(a["MTRANS"])
print(a)
```

```python
#data split
x=a.iloc[:,:-1]
y=a["Obesity"]
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
ypred=model.predict(x_test)
```

```python
print(ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor
b=DecisionTreeRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```

```python
print(ypred)
```

```python
from sklearn.ensemble import RandomForestRegressor
b=RandomForestRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```
