```python
from sklearn.preprocessing import LabelEncoder
b=LabelEncoder()
a["Location"]=b.fit_transform(a["Location"])
a["Type"]=b.fit_transform(a["Type"])
a["Bathrooms"]=b.fit_transform(a["Bathrooms"])
a["Bedrooms"]=b.fit_transform(a["Bedrooms"])
print(a)
```

```python
data=a.corr()
sns.heatmap(data, annot=True)
```

```python
sns.pairplot(data=a)
```

```python
#data split
x=a.iloc[:,:-1]
y=a["Price"]
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
ypred=model.predict(x_test)
```

```python
print(ypred)
```

```python
print(model.predict([[162,4,14,1,140]]))
```

```python
from sklearn.tree import DecisionTreeRegressor
b=DecisionTreeRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```

```python
print(ypred)
```

```python
print(b.predict([[162,4,14,1,140]]))
```

```python
from sklearn.ensemble import RandomForestRegressor
b=RandomForestRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```

```python
print(ypred)
```

```python
print(b.predict([[162,4,14,1,140]]))
```
