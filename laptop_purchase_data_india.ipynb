```python

#Transform the values to numbers
from sklearn.preprocessing import LabelEncoder
b=LabelEncoder()
a["Customer_ID"]=b.fit_transform(a["Customer_ID"])
a["Gender"]=b.fit_transform(a["Gender"])
a["City"]=b.fit_transform(a["City"])
a["Income_Level"]=b.fit_transform(a["Income_Level"])
a["Purchase_Mode"]=b.fit_transform(a["Purchase_Mode"])
a["Brand_Purchased"]=b.fit_transform(a["Brand_Purchased"])
a["Price_Range"]=b.fit_transform(a["Price_Range"])
a["Payment_Mode"]=b.fit_transform(a["Payment_Mode"])
a["Purchase_Frequency"]=b.fit_transform(a["Purchase_Frequency"])
print(a)
```

```python
#data split
x=a.iloc[:,:-1]
y=a["Satisfaction_Rating"]
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
print(x_test)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
ypred=model.predict(x_test)
```

```python
print(ypred)
```

```python
from sklearn.tree import DecisionTreeRegressor
b=DecisionTreeRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```

```python
print(ypred)
```

```python
from sklearn.ensemble import RandomForestRegressor
b=RandomForestRegressor()
b.fit(x_train,y_train)
ypred=b.predict(x_test)
```
